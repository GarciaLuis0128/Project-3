```python

import pandas as pd
import matplotlib.pyplot as plt
# 
#Joy:Most Popular job and how does this compare by income?
#How does the Job category and salary compare by country? 


```

```python
csv_file_path = "C:/Users/Joyjo/Downloads/jobs_in_data.csv"
```

```python
# DataFrame from the CSV file
df = pd.read_csv(csv_file_path)
#print dataframe
df
```

```python
# Clean Dataframe by dropping unnecessary columns ('salary_currency' and 'salary_in_usd')
clean_df = df.drop(columns=['salary_currency', 'salary_in_usd'])
clean_df
```

```python
job_income = df.groupby('job_category')['salary'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
job_income.plot(kind='bar', color='skyblue')
plt.title('Average Income by Job')
plt.xlabel('Job')
plt.ylabel('Average Income')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
cross_table = pd.crosstab(df['job_category'], df['company_size'])
plt.figure(figsize=(12, 8))
cross_table.plot(kind='bar', stacked=True, cmap='viridis')
plt.title('Job Category vs Company Size')
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Company Size', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
